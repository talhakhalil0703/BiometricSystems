Considering a variable x, with interval $x \in (-6, 6) $, we create the following function:

$$f(x)=sin x + \frac{cos(2x)}{2} + \frac{sin(3x)}{3} + \frac{cos(4x)}{4}$$

And then plot it in various styles.

```python
# Importing needed libraries
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-6, 6) # Generating x with interval -6 to 6
y = np.sin(x) + (np.cos(2*x)/2 ) + (np.sin(3*x)/3 ) + (np.cos(4*x)/4 ) # Calculating f(x)
```

Plotting with green line, carrot markers, and a straight line.

```python
plt.plot(x,y, '-g^')
plt.title('$f(x)=sin x + \\frac{cos(2x)}{2} + \\frac{sin(3x)}{3} + \\frac{cos(4x)}{4}$')
plt.xlabel('x')
plt.ylabel('f(x)')
```

Plotting with red line, o markers, and a dashed line.

```python
plt.plot(x,y, '--ro')
plt.title('$f(x)=sin x + \\frac{cos(2x)}{2} + \\frac{sin(3x)}{3} + \\frac{cos(4x)}{4}$')
plt.xlabel('x')
plt.ylabel('f(x)')
```

We can see the difference in plotting; where the first is plotted using a green straight line and carrot markers and the second plot is plotted using a red dashed line and o markers.

We first must expand the age vector by making use of it's indices to calculate the mean and standard deviation. From which we can then calculate the normal distribution.

```python
# We first expand the intial age vector. 
expanded_age_vector = np.array(0)
for index, age_value in enumerate(age):
    expanded_age_vector = np.append(expanded_age_vector, np.repeat(index, age_value))

#Now we calculate the mean and the standard deviation.
age_mu = np.mean(expanded_age_vector)
age_std = np.std(expanded_age_vector)
print('Age_mu: ', age_mu)
print('Age_std: ', age_std)
```

We can proceed to now plot the normal distribution as described by the following equation:
$$f(x)=\frac{1}{\sigma \sqrt{2\pi}}e^{-\frac{1}{2} (\frac{x-\mu}{\sigma})^2}$$


> Note for smoothing purposes we use a graping_age variable to create a smooth bell curve.

```python
from scipy.stats import norm
graphing_age = np.linspace(0,len(age)) # For smooth bell curve
```

Here is the graph of the normal distribution with the indices changed to reflect the age range.

```python
plt.plot(graphing_age, norm.pdf(graphing_age, age_mu, age_std))
plt.xlabel('Age')
plt.ylabel('Probability')
tick_age = np.linspace(0,len(age),len(age)) # Add labels for the age group

plt.xticks(tick_age, labels=age_labels, rotation=45)
```

Here is the graph of the normal distribution with the original indices without the labels.

```python
plt.plot(graphing_age, norm.pdf(graphing_age, age_mu, age_std))
plt.xlabel('Age Index')
plt.ylabel('Probability')
```

We can conclude that the mean age of the population was 5.33 in indices; as in  age group 50-59. With a standard deviation of 1.85 in indices.
